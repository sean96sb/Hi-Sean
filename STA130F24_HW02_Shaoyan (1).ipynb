````python
import numpy as np

def monte_hall_sim(reps, swap):
    """
    Simulates the Monte Hall problem to test 
    whether switching doors improves the odds of winning.

    Args:
        reps (int): The number of times to run the simulation.
        swap (bool): True to simulate the "swap strategy", 
                     False to simulate keeping the initial choice. 

    Returns:
        float: The estimated probability of winning based on the simulation.
    """
    wins = 0
    for _ in range(reps):
        winning_door = np.random.choice([1-3])
        player_choice = np.random.choice([1-3])

        if swap:
            # Reveal a losing door that's not the player's choice
            revealed_door = np.random.choice([door for door in [1-3] if door != winning_door and door != player_choice])
            
            # Switch player's choice to the remaining unopened door
            player_choice = next(door for door in [1-3] if door != revealed_door and door != player_choice)

        if player_choice == winning_door:
            wins += 1

    return wins / reps
````

    # If the word is a character, store it as the current character

    word_pair = word + ' ' + words[i + 1]  # Form the word pair

    # Update the word pair frequency for the character
    word_used2C[character][word_pair] += 1

    # Update the next word transition from the word pair

3. **Efficient loop structure:** The `for` loop iterates over the `words` list, forming word pairs (`word` and `words[i+1]`) and checking transitions to the next word (`words[i+2]`). This reduces redundancy and simplifies the code logic.

````python
# Markovian Chatbot Extension #2

from collections import Counter, defaultdict
# `avatar` is a dataset, and `character` is one of it's columns
characters = Counter("\n"+ avatar.character.str.upper().str.replace(' ','.')+":")
# this code changes the type of the `character` column to `str`; then,
# makes the text uppercase, and replaces spaces with '.'

nested_dict = lambda: defaultdict(nested_dict)
word_used2C = nested_dict()
next_word2C = nested_dict()

for i,word in enumerate(words[:-2]):
    if word in characters:
        character = word
        
    if character not in word_used2C:
        word_used2C[character] = dict()
    if word+' '+words[i+1] not in word_used2C[character]:
        word_used2C[character][word+' '+words[i+1]] = 0
    word_used2C[character][word+' '+words[i+1]] += 1
    
    if character not in next_word2C:
        next_word2C[character] = dict()
    if word+' '+words[i+1] not in next_word2C[character]:
        next_word2C[character][word+' '+words[i+1]] = dict()
    if words[i+2] not in next_word2C[character][word+' '+words[i+1]]:
        next_word2C[character][word+' '+words[i+1]][words[i+2]] = 0
    next_word2C[character][word+' '+words[i+1]][words[i+2]] += 1
````
